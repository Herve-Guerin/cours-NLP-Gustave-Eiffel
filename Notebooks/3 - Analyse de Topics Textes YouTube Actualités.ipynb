```
# 2.e
# répartition un peu arbitraire entre catégories vides et pleines
# les premières faisant office de stopwords, dans un modèle sacs de mots
CATEGORIES_VIDES = ['ADP', 'AUX', 'CONJ', 'CCONJ', 'DET', 'PART', 'PRON',
                    'PUNCT', 'SCONJ', 'SYM', 'SPACE']
CATEGORIES_PLEINES = ['ADJ', 'ADV', 'INTJ', 'NOUN', 'NUM', 'PROPN', 'VERB', 'X']
```

```
# 2.f
# cette fonction servira à la fois sur les analyses déjà faites et sur le texte de référence utilisé à fins de comparaisons
def selectionne_lemme(lemme, pos, est_stopword):
  selection_lemme = ''
  if pos in CATEGORIES_PLEINES and est_stopword == False:
    selection_lemme = lemme
  return selection_lemme
```

```
# 2.g
# utilisation sur les analyses déjà faites, création d'une nouvelle colonne, en utilisant la fonction suivante,
# qui prend en argument la liste des tokens résultat de l'analyse nlp spacy effectuée auparavant
def cree_l_items_sur_resultat_nlp(l_tokens_nlp, retour='lemme'):
  l_items = []
  for token in l_tokens_nlp:
    lemme = selectionne_lemme(token[3], token[4], token[7])
    if lemme != '':
      if retour == 'lemme': l_items.append(lemme)
      if retour == 'tout': l_items.append(token)
  return l_items
```

### Création du modèle de topics NMF

###### Nous utilisons la librairie scikit-learn (pb avec gensim)
##### Création des sacs de mots, des dictionnaires index -> termes et termes -> index
##### Il vaut mieux partir d'une représentation tf-idf pour le modèle NMF 

```
# 3.a
# tokeniseur rudimentaire, car tout a été préparé en amont, sert pour remplacer le tokeniseur par défaut de scikit-learn
def simple_split(s):
  return s.split()
```

```
# 3.b
# sous forme d'une fonction avec comme argument principal l_textes, liste des textes de départ, s'obtenant via df[TEXTES].values.tolist()
# plus paramètres : 
# max_dfreq: pour enlever les termes qui sont trop fréquents, présents dans trop de documents, défaut skearn 1.0 (100 %, présents partout ne suffit alors pas pour être éliminé 
# min_dfreq: pour enlever les termes qui sont trop rares, présents dans pas assez de documents, défaut sklearn 1 (nb docs < 1, présents dans 1 seul doc suffit pour être conservé),
# à modifier pour larges corpus (le défaut gensim est à 5)
# nb_total_termes: nb total de termes retenus, selon la fréquence, par défaut (None) on garde tout, une bonne fourchette est entre 10000 et 15000 termes
# il faut aussi préciser le tokenisateur, ici, vu la préparation effectuée en amont, le texte est une liste des tokens séparés par des blancs,
# un simple split suffit
def cree_sac_de_mots_TFIDF(l_textes, max_dfreq, min_dfreq, nb_total_termes):

  # objet pour vectoriser les textes
  tfidf = TfidfVectorizer(max_df=max_dfreq, min_df=min_dfreq, max_features=nb_total_termes, 
                          lowercase=False, tokenizer=simple_split)

  # vectorisation, matrice docs X termes          
  mx_docs_termes = tfidf.fit_transform(l_textes)
  ix_termes = tfidf.get_feature_names()
  nb_termes = len(ix_termes)

  return mx_docs_termes, ix_termes, nb_termes
```

```
# 3.c
mx_docs_termes, ix_termes, nb_termes = cree_sac_de_mots_TFIDF(l_textes, 0.95, 5, 10000)
# vérification
print(nb_termes)
```

```
# 3.d
# on pourra aussi avoir besoin du dictionnaire de termes inversé (termes -> index)
def cree_termes_ix(ix_termes):
  termes_ix = {}
  for no_terme in range(len(ix_termes)):
    terme = ix_termes[no_terme]
    termes_ix[terme] = no_terme
  return termes_ix
```

#### Créer le modèle de topic NMF proprement dit

```
# 3.f
# sous forme d'une fonction avec comme arguments le nb de topics et la matrice docs X termes (TFIDF)
def cree_modele_topics_NMF(nb_topics, mx_docs_termes):
  
  # utilisation de la fonction scikit-learn
  nmf = NMF(n_components=nb_topics, random_state=1).fit(mx_docs_termes)
        
  # matrice docs X topics  
  mx_docs_topics = nmf.fit_transform(mx_docs_termes)
  # matrice topics X termes
  mx_topics_termes = nmf.components_ 
  # erreur de reconstruction (pas très pertinente (plus il y a de topics, plus elle est faible), mais cependant conservée)
  nmf_reconstruction_err = nmf.reconstruction_err_
        
  return mx_docs_topics, mx_topics_termes, nmf_reconstruction_err
```

##### Fonctions annexes et utiles construites autour des matrices de topics

```
# 3.h
# trouver la liste des top termes (plus leurs poids) d'un topic
def topic_top_termes(mx_topics_termes, ix_termes, no_topic, nb_top_termes):
  v_topic_termes = mx_topics_termes[no_topic]    
  return [(ix_termes[no_terme], v_topic_termes[no_terme])\
          for no_terme in np.argsort(v_topic_termes)[:-nb_top_termes-1:-1]]
```

```
# 3.i
# trouver la liste des top docs (plus leurs poids) d'un topic
def topic_top_docs(mx_docs_topics, no_topic, nb_top_docs):
  v_topic_docs = mx_docs_topics[:,no_topic]    
  return [(no_doc, v_topic_docs[no_doc])\
          for no_doc in np.argsort(v_topic_docs)[:-nb_top_docs-1:-1]]
```

```
# 3.j
# trouver la masse totale sur les docs d'un topic
def topic_masse_totale(mx_docs_topics, no_topic):
  return mx_docs_topics[:,no_topic].sum() 
```

```
# 3.k
# dériver certaines informations (nommage + pesage) sur le modèle de topics
def informations_derivees(mx_docs_topics, mx_topics_termes):

  # liste des topics avec un nom parlant dérivé des noms des termes les plus représentatifs,
  # plus le poids des topics 
  nb_topics = mx_topics_termes.shape[0]
  l_topics = []
  for no_topic in range(nb_topics):
    topic_attributs = {}
    top_2_termes = topic_top_termes(mx_topics_termes, ix_termes, no_topic, 2)
    topic_attributs['nom'] = "%s %s" % (top_2_termes[0][0], top_2_termes[1][0])
    topic_attributs['masse'] = topic_masse_totale(mx_docs_topics, no_topic)
    l_topics.append(topic_attributs)

  return l_topics
```

```
# 4.e
# en préparatoire, il faut préparer une version normalisée de la matrice docs X topics, 
# telle que pour chaque document, la somme des poids des topics sur le document fait 1
# plus un vecteur avec pour chaque document sa longueur en nombre de tokens
# plus un vecteur donnant la fréquence sur le corpus pour chaque terme du dictionnaire
```

```
# 4.f
# version normalisée de la matrice docs X topics, (comme pour LDA !)
# telle que pour chaque document, la somme des poids des topics sur le document fait 1
def cree_mx_norm_docs_topics(mx_docs_topics):
  mx_norm_docs_topics = np.zeros(mx_docs_topics.shape)
  for no_doc in range(mx_docs_topics.shape[0]):
    sum_on_doc = mx_docs_topics[no_doc,:].sum()
    if sum_on_doc == 0.0: sum_on_doc = 1.0
    mx_norm_docs_topics[no_doc,:] = mx_docs_topics[no_doc,:] / sum_on_doc
  return mx_norm_docs_topics
```

```
# 4.g
# vecteur avec pour chaque document sa longueur en nombre de tokens (qu'ils soient dans ou hors du dictionnaire)
def cree_v_docs_nb_tokens(l_textes_l_tokens):
  nb_docs = len(l_textes_l_tokens)
  v_docs_nb_tokens = np.zeros((nb_docs), dtype = int)
  for no_doc in range(nb_docs):
    v_docs_nb_tokens[no_doc] = len(l_textes_l_tokens[no_doc])
  return v_docs_nb_tokens
```

```
# 4.h
# vecteur donnant la fréquence sur le corpus pour chaque terme du dictionnaire
def cree_v_freqs_termes(l_textes_l_tokens, termes_ix):
  nb_termes = len(termes_ix)
  v_freqs_termes = np.zeros((nb_termes), dtype = int)
  for l_tokens in l_textes_l_tokens:
    for token in l_tokens:
      if token in termes_ix: 
          v_freqs_termes[termes_ix[token]] += 1
  return v_freqs_termes
```

```
# 5.a
# Evaluation de la cohérence sémantique LCP, 
# topic par topic, puis moyenne sur les topics
def evalue_lcp_coherence_semantique(mx_docs_topics, mx_topics_termes, ix_termes, termes_ix, l_textes_l_tokens, 
                                    nb_top_termes, taille_echanti=5000):
        
    # obtenir la liste topic par topic de tous les tops termes 
    nb_topics = mx_topics_termes.shape[0]
    l_topics_top_termes = []
    for no_topic in range(nb_topics):
        l_topics_top_termes.append(topic_top_termes(mx_topics_termes, ix_termes, no_topic, nb_top_termes))
        
    # avoir une liste des indices sur lesquels effectuer les calculs (assez coûteux)
    # la valeur correcte dépends surtout du nombre de top termes (au carré)
    nb_docs = mx_docs_topics.shape[0] 
    taille_echanti = min([nb_docs, taille_echanti])
    l_echanti_ix = sample(list(range(nb_docs)), taille_echanti)
    # nouvelle valeur (provisoire) du nombre de docs
    nb_docs = len(l_echanti_ix)
        
    # on teste les co-occurrences de 2 termes dans le même texte, et donc pour chaque texte (de l'échantillon)
    # on crée un dictionnaire disant si le terme est dans le texte ou non
    l_d_texte_termes = []
    for no_doc in l_echanti_ix:
        d_texte_termes = {}
        # utilisant la list des termes du texte
        l_tokens = l_textes_l_tokens[no_doc]
        for token in l_tokens:
            # ne considérer just use the lexunits from the dictionary
            # (one could even use only the lexunits in any one of the top lexical units for the topics)
            if token in termes_ix:
                if token not in d_texte_termes: d_texte_termes[token] = 1
        l_d_texte_termes.append(d_texte_termes)
        
    # calcul proprement dit
    modele_lcp_coherence = 0.0
    l_topics_lcp_coherences = []
    for no_topic, top_termes in enumerate(l_topics_top_termes):
        topic_lcp_coherence = 0.0
        for i in range(1, nb_top_termes):
            terme_i = top_termes[i][0]
            contrib_i = 0.0
            for j in range(0, i):
                terme_j = top_termes[j][0]
                nb_docs_avec_j = 0
                nb_docs_avec_i_et_j = 0
                for d_texte_termes in l_d_texte_termes:
                    if terme_j in d_texte_termes:
                        nb_docs_avec_j += 1
                        if terme_i in d_texte_termes:
                            nb_docs_avec_i_et_j += 1
                        
                # terme correctif original pour éviter log de 0 était 1
                # mais une plus petite valeur donne de 
                if nb_docs_avec_j > 0:
                    contrib_i_j = math.log(((nb_docs_avec_i_et_j / nb_docs) + 1e-12) \
                                           / (nb_docs_avec_j / nb_docs))
                else:
                    contrib_i_j = 0.0
                contrib_i += contrib_i_j 
            topic_lcp_coherence +=  contrib_i    
                    
        l_topics_lcp_coherences.append(topic_lcp_coherence)
        modele_lcp_coherence += topic_lcp_coherence
        #print("COH SEM LCP Topic %d" % no_topic, end="\r")
            
    modele_lcp_coherence /= nb_topics
            
    # résultats
    return modele_lcp_coherence, l_topics_lcp_coherences
```

###### CoherenceModel avec coherence = 'u_mass', 'c_v', 'c_uci', 'c_npmi'

- c_uci : cohérence Pointwise Mutual Information
- c_npmi : c_pmi améliorée

```
# 5.c
# encapsuler la fonction de calcul c_v gensim
def evalue_c_v_coherence_semantique(mx_topics_termes, l_textes_l_tokens, nb_top_termes, taille_fenetre=110, taille_echanti=5000):
  
  # obtenir la liste topic par topic de tous les tops termes 
  nb_topics = mx_topics_termes.shape[0]
  l_topics_top_termes = []
  for no_topic in range(nb_topics):
      # la fonction topic_top_termes renvoie et le terme et le poids du terme dans le topic, ne garder que le premier
      l_top_termes = [topic_top_terme[0] for topic_top_terme in topic_top_termes(mx_topics_termes, ix_termes, no_topic, nb_top_termes)]
      l_topics_top_termes.append(l_top_termes)

  # la liste des textes est déjà là, mais en prendre un échantillon si pb de performances
  nb_docs = len(l_textes_l_tokens)
  taille_echanti = min([nb_docs, taille_echanti])
  l_echanti_ix = sample(list(range(nb_docs)), taille_echanti)
  l_textes_l_tokens_echanti = [l_textes_l_tokens[ix] for ix in l_echanti_ix]

  # le dictionnaire attendu est un dictionnaire gensim
  # le créer à partir des textes de départ
  dict_gensim = corpora.Dictionary(l_textes_l_tokens)
  
  # pour c_v, la taille de la fenêtre d'examen est par défaut de 110 (10 pour pmi et npmi)
  # comme nous avons éliminé les stop-words, on peut réduire cette taille à 60
  cm_c_v = CoherenceModel(topics=l_topics_top_termes, texts=l_textes_l_tokens_echanti, dictionary=dict_gensim, coherence='c_v', window_size=taille_fenetre)
  modele_c_v_coherence = cm_c_v.get_coherence() 
  # pour info, on peut aussi obtenir la liste des cohérences topics par topics
  l_topics_c_v_coherences = cm_c_v.get_coherence_per_topic()

  return modele_c_v_coherence, l_topics_c_v_coherences
  
```

```
# 5.e
# Evaluation de la métrique FREX, 
# topic par topic, puis moyenne sur les topics
# adaptation du code R du package STM
def evalue_FREX(mx_topics_termes, nb_top_termes, poids_exclusivite=0.7):

    nb_topics = mx_topics_termes.shape[0]

    # transposition de la matrice topics X termes, d'où une matrice termes X topics
    H = mx_topics_termes
    tH = np.transpose(H)
    
    # somme sur chaque topic on each topics of the lexunits weights
    masses_topics = np.sum(tH, axis=1)
    # matrice des distributions de termes, normalisée à 1 topic par topic (distribution proba)
    # servira pour les exclusivités
    norm_tH = tH / masses_topics[:, None]
    
    # exclusivité
    # tri des termes sur les topics (colonnes), y mettre les nos d'ordre des termes, topic par topic
    # le faire colonne par colonne 
    X_exclu = np.zeros((norm_tH.shape[0], norm_tH.shape[1]))
    for c in range(norm_tH.shape[1]):
        V = norm_tH[:,c]
        SV = ss.rankdata(V, method='average')
        X_exclu[:,c] = SV
    # et diviser par le nb de lignes (termes)
    mx_exclu = X_exclu / norm_tH.shape[0]
    
    # fréquence
    # tri des termes sur les topics (colonnes), y mettre les nos d'ordre des termes, topic par topic
    # le faire colonne par colonne
    X_freq = np.zeros((tH.shape[0], tH.shape[1]))
    for c in range(tH.shape[1]):
        V = tH[:,c]
        RV = ss.rankdata(V, method='average')
        X_freq[:,c] = RV
    # et diviser par le nb de lignes (termes)
    mx_freq = X_freq / tH.shape[0]

    # moyenne harmonique
    mx_frex = 1/(poids_exclusivite/mx_exclu + (1-poids_exclusivite)/mx_freq)
    
    # indices à appliquer (sur la matrice (transposée) topics X termes))
    idx_termes = np.zeros((tH.shape[0], tH.shape[1]))
    for c in range(tH.shape[1]):
        V = tH[:,c]
        SV = np.argsort(V)
        FV = np.flip(SV)
        idx_termes[:,c] = FV
        
    # garder les top termes
    idx_top_termes = idx_termes[range(nb_top_termes),:]
    # avec des indices entiers
    idx_top_termes = idx_top_termes.astype(np.int)
    
    # list des frex pour tous les topics
    l_topics_frex = [] 
    for no_topic in range(nb_topics):
        topic_frex = mx_frex[idx_top_termes[:,no_topic],no_topic].sum()
        l_topics_frex.append(topic_frex)
  
    # et moyenne générale
    modele_frex = sum(l_topics_frex) / nb_topics
        
    # return results
    return modele_frex, l_topics_frex
```

```
# 5.g
# évaluation inter-topic des distributions des top termes sur les topics, avec pénalisation si dispersion des termes sur plusieurs topics
# sur plusieurs topics en groupe

# fonction auxiliaire donnant un coefficient de puissance sur les nb de topics sur lequels un terme est présent
def calcule_coeff_dispersion(nb_topics):
    
    if nb_topics == 2: alpha = 1.0
    if nb_topics >= 25: alpha = 0.5
        
    if nb_topics >= 3 and nb_topics <= 24:
        l_inf = 0.5
        l_sup = 0.67
        alpha = l_inf
        y = alpha * (1 - alpha ** (nb_topics-1)) / (1 - alpha)
        while abs(y - 1) >= 0.0000001:
            if y > 1: 
                l_sup = alpha
            if y <= 1:
                l_inf = alpha
            alpha = (l_inf + l_sup) / 2
            y = alpha * (1 - alpha ** (nb_topics-1)) / (1 - alpha)
    return alpha


# fonction d'évaluation de la dispersion (en fait on renvoie une concentration)
def evalue_concentration_top_termes(mx_topics_termes, ix_termes, nb_top_termes):
        
    # obtenir la liste topic par topic de tous les tops termes 
    nb_topics = mx_topics_termes.shape[0]
    l_topics_top_termes = []
    for no_topic in range(nb_topics):
        l_topics_top_termes.append(topic_top_termes(mx_topics_termes, ix_termes, no_topic, nb_top_termes))
    
    # avoir un dictionnaire comptant le nb de topics pour chaque top termes
    d_top_termes_nb_topics = {}
    for no_topic, top_termes in enumerate(l_topics_top_termes):
        for (terme, poids) in top_termes:
            if terme not in d_top_termes_nb_topics:
                d_top_termes_nb_topics[terme] = 1
            else:
                d_top_termes_nb_topics[terme] += 1
        
    # sur chaque terme faire le calcul de concentration
    concentration_top_termes = 0.0
    alpha = calcule_coeff_dispersion(nb_topics)
    for terme, nb_topics_terme in d_top_termes_nb_topics.items():
        if nb_topics_terme == 1:
            val_terme = 1.0
        elif nb_topics_terme == nb_topics:
            val_terme = 0.0
        else:
            val_terme = 1 - (alpha * (1 - alpha ** (nb_topics_terme-1)) / (1 - alpha))
        concentration_top_termes += val_terme
            
    # prendre la moyenne, en considérant le nb de combinaisons possibles (nb_top_termes * nb_topics) 
    # plutôt que le nb distinct de termes
    concentration_top_termes /= (nb_topics * nb_top_termes)      
   
    # return results
    return concentration_top_termes
```

```
# 5.m
# visualisation des évaluations selon le nb de topics

def montre_1_methode_N_metrics_evals(l_nb_topics, l_metriques, l_l_evaluations, 
                                     nom_methode, nb_top_termes, sauvegarde=True):
    
    # on trace N lignes, une par métrique d'évaluation,
    # avec nb de topics en x, résultats d'évaluation (sur l'ensemble des topics) en y, 
    # les résultat étant réduits à une échelle entre 0 et 1 pour chaque métrique

    nb_metriques = len(l_metriques)    
    # préparation des données
    l_l_evals_reduites = []
    for no_metrique in range(nb_metriques):
      l_evals_metrique = l_l_evaluations[no_metrique]
      eval_min = min(l_evals_metrique)
      eval_max = max(l_evals_metrique)
      l_evals_reduites = [(eval - eval_min) / (eval_max - eval_min)\
                         for eval in l_evals_metrique]
      l_l_evals_reduites.append(l_evals_reduites)

    # graphiques
    fig = go.Figure()
    # boucle sur chaque métrique
    for no_metrique in range(nb_metriques):
      nom_metrique = l_metriques[no_metrique]
      l_evals_reduites = l_l_evals_reduites[no_metrique]
      fig.add_trace(go.Scatter(x=l_nb_topics, y=l_evals_reduites,
                                  mode='lines+markers', name=nom_metrique))
        
    # items péripheriques      
    # titre
    titre = "Evaluations pour la méthode %s (sur les %d top termes)" % (nom_methode, nb_top_termes)
    # légende des x
    titre_x = "Nb de topics"
    fig.update_layout(title=titre, xaxis_title=titre_x)
        
    # sauver éventuellement
    if sauvegarde == True:
      nom_fichier = "%s.png" % titre
      fig.write_image(nom_fichier)
      files.download(nom_fichier)

    fig.show()
```

#### Création du modèle de topics LDA

###### Nous utilisons la librairie gensim
##### Création des sacs de mots, des dictionnaires index -> termes et termes -> index
##### Il vaut mieux partir d'une représentation fréquence absolue pour le modèle LDA 

```
# 6.a
# sous forme d'une fonction avec comme argument principal l_textes_tokens, liste des textes (texte = listes de tokens) de départ
# plus paramètres : 
# max_dfreq: pour enlever les termes qui sont trop fréquents, présents dans trop de documents, défaut gensim 0.5 
# (50 %, beaucoup trop discriminatif, surtout quand le sujet a été précisé) 
# min_dfreq: pour enlever les termes qui sont trop rares, présents dans pas assez de documents, défaut gensim 5 (nb docs < 5, présents dans moins de 5 docs),
# à modifier pour larges corpus
# nb_total_termes: nb total de termes retenus, selon la fréquence, par défaut (100000) on garde beaucoup, None, on garde tout, 
# une bonne fourchette est entre 10000 et 15000 termes
# min_freq_rang : pour éliminer les N termes les plus fréquents,
# on les voit en général polluer beaucoup de topics (commencer par 2, puis étendre au fur et à mesure)
def cree_sac_de_mots_FREQ(l_textes_l_tokens, max_dfreq, min_dfreq, nb_total_termes, min_freq_rang):

  # création du dictionnaire (sous forme gensim)
  ix_termes = corpora.Dictionary(l_textes_l_tokens)
  # filtrer les extrêmes, termes trop ou trop peu fréquents
  ix_termes.filter_extremes(no_below=min_dfreq, no_above=max_dfreq, keep_n=nb_total_termes)
  # aussi enlever les N termes les plus frequents (en général les 2 plus fréquents)
  ix_termes.filter_n_most_frequent(min_freq_rang)
  nb_termes = len(ix_termes)

  # vectorisation, création de la matrice doc X termes (c'est là une liste)
  mx_docs_termes = [ix_termes.doc2bow(l_tokens) for l_tokens in l_textes_l_tokens]  

  return mx_docs_termes, ix_termes, nb_termes
```

```
# 6.b
mx_docs_termes, ix_termes, nb_termes = cree_sac_de_mots_FREQ(l_textes_l_tokens, 0.5, 5, 10000, 10)
```

```
# 6.c
# sous forme d'une fonction avec comme arguments le nb de topics et la matrice docs X termes (fréquences simples)
# ainsi que le dictionnaire index -> termes
# plus paramètres : 
# eta (beta ds littérature), controlant la distribution des termes sur les topics
# par défaut None, recommandé : 0.01
# alpha, controlant la distribution des topics sur les documents
# par défaut 'symmetric', recommandé : 1 (divisé dans la fonction par / nb topics)
def cree_modele_topics_LDA(nb_topics, mx_docs_termes, ix_termes, eta=0.01, alpha=1):

  if alpha not in ['auto', 'symmetric', 'asymmetric']:
    alpha = alpha / nb_topics
  lda = LdaModel(mx_docs_termes, num_topics=nb_topics, id2word=ix_termes,
                 eta=eta, alpha=alpha)
        
  # gensim créée une liste de listes tant pour la matrice docs X topics que pour la matrice topics X termes
  # créer la matrice numpy mx_docs_topics à partir de la liste de nb_docs listes de nb_topics poids
  nb_docs = len(mx_docs_termes)
  mx_docs_topics = np.zeros((nb_docs, nb_topics))
  # récupération de la matrice docs X termes créée par gensim (une liste)
  l_docs_topics = lda[mx_docs_termes]
  for no_doc, l_doc_topics in enumerate(l_docs_topics):
    l_doc_topics.sort(key=lambda x: x[0]) # pour être sûr
    for doc_topic in l_doc_topics:
      mx_docs_topics[no_doc, doc_topic[0]] = doc_topic[1]
    
  # pour mx_topics_termes, pour chaque topic trouver les poids pour les termes
  nb_termes = len(ix_termes)
  mx_topics_termes = np.zeros((nb_topics, nb_termes))
  for no_topic in range(nb_topics):
    l_topic_termes = lda.get_topic_terms(no_topic, topn=nb_termes)
    # cette liste de (int, float) doit être triée selon l'index (int) du dictionnaire
    l_topic_termes.sort(key=lambda x: x[0])
    for topic_terme in l_topic_termes:
      mx_topics_termes[no_topic, topic_terme[0]] = topic_terme[1]

  return mx_docs_topics, mx_topics_termes
```

```
# 6.d obtention du modèle
mx_docs_topics, mx_topics_termes = cree_modele_topics_LDA(12, mx_docs_termes, ix_termes, eta=0.001, alpha='asymmetric')

print(mx_docs_topics.shape)
print(mx_topics_termes.shape)
```

##### Si la répartition des termes sur les topics est par trop dispersée, on peut la concentrer de façon à diminuer l'influence des termes présents sur trop de topics 
##### La formule utilisée est quelque peu analogue à la formule du tf-idf avec sa pondération appliquée au terme de base tf

```
# 6.g fonction de calcul de la matrice topics X termes pondérée
def calcule_mx_topics_termes_ponderes(mx_topics_termes):
  mx_topics_termes_ponderes = np.zeros(mx_topics_termes.shape)
  nb_topics = mx_topics_termes.shape[0]
  nb_termes = mx_topics_termes.shape[1]
  for no_terme in range(nb_termes):
    log_sur_terme = 0.0
    for no_topic in range (nb_topics):
      log_sur_terme += math.log(mx_topics_termes[no_topic, no_terme] + 1e-12) / nb_topics
      for no_topic in range (nb_topics):
        mx_topics_termes_ponderes[no_topic, no_terme] = mx_topics_termes[no_topic, no_terme] \
        * (math.log(mx_topics_termes[no_topic, no_terme] + 1e-12) - log_sur_terme)
  return  mx_topics_termes_ponderes
```

### Phase 1) Obtention des groupes nominaux

```
# 7.g dependency matcher de spaCy
matcher = DependencyMatcher(nlp.vocab)
```

```
# 7.h définition de ce qu'est un groupe nominal assez simple et intéressant (et en français)
# groupes nominaux à 2 et 3 éléments
pattern_2 = [
  {
    "RIGHT_ID": "nom_tête",       
    "RIGHT_ATTRS": {"POS": "NOUN"}  
  },
  {
    "LEFT_ID": "nom_tête",
    "REL_OP": ">",        
    "RIGHT_ID": "dépendance",
    "RIGHT_ATTRS": {"DEP": {"IN": ["amod", "nmod"]}},    
  }
]

pattern_3 = [
  {
    "RIGHT_ID": "nom_tête",       
    "RIGHT_ATTRS": {"POS": "NOUN"}  
  },
  {
    "LEFT_ID": "nom_tête",
    "REL_OP": ">",        
    "RIGHT_ID": "dépendance_1",
    "RIGHT_ATTRS": {"DEP": {"IN": ["amod", "nmod"]}},    
  },
  {
    "LEFT_ID": "dépendance_1",
    "REL_OP": ">",        
    "RIGHT_ID": "dépendance_2",
    "RIGHT_ATTRS": {"DEP": {"IN": ["amod", "nmod"]}},    
  }
]
```

```
# 7.i le matcher qu'on va utiliser, nommé "GN" comme groupe nominal, et 2 et 3 selon nombre de dépendances
matcher.add("GN_2", [pattern_2])
matcher.add("GN_3", [pattern_3])
```

```
# 7.j 
# fonction qui pour chaque texte, renvoie les groupes nominaux, 
# sous la forme d'une liste de listes des indices des tokens participant aux groupes nominaux
# cette analyse est faite ici en parallèle (et en double pour l'appel de nlp !) 
# mais devrait être conduite en même temps que l'analyse nlp !!! 
def groupes_nominaux_par_nlp(texte, nlp, matcher):

  l_groupes_nominaux = []

  # analyse spacy
  doc = nlp(texte)

  # langue du texte
  langue = doc._.language['language']

  # récupération des groupes nominaux
  if langue == 'fr':
    matches = matcher(doc)

    # un match est une liste de deux termes, un identifiant et la liste des index de tokens
    for x_groupe_nominal in matches:
      l_groupes_nominaux.append(x_groupe_nominal[1])

  return l_groupes_nominaux
```

```
# 7.k
# plus que la liste des indices des tokens des groupes nominaux, ce qui nous intéresse c'est un texte équivalent
# 3 possibilités de reconstruction :
# - concaténation des lemmes (dans l'ordre où ils apparaissent dans le texte !) des tokens repérés au sein du groupe
# - concaténation des formes d'origine (toujours dans le bon ordre) de ces tokens
# - tout le texte entre le début du premier token et la fin du dernier token
# on récupère ici les 3 formes, pour montrer les différences 

def reconstruit_groupe_nominal(groupe_nominal, l_tkn_sous_titres, txt_sous_titres):

  groupe_lemmes = ""
  groupe_tokens = ""
  groupe_texte = ""
    
  # le groupe nominal est une liste d'index qu'il convient de trier dans l'ordre linéaire
  groupe_nominal.sort()
  
  # procéder différemment selon la nature de la récupération, d'abord lemmes et tokens
  for ix, i_tkn in enumerate(groupe_nominal):
    tkn = l_tkn_sous_titres[i_tkn]
    texte_lemme = tkn[3]
    texte_token = tkn[0]
    groupe_lemmes += texte_lemme
    groupe_tokens += texte_token
    # mettre le séparateur '_', sauf si en dernier
    if ix < len(groupe_nominal)-1: 
      groupe_lemmes += '_'
      groupe_tokens += '_'

  # pour récupération du début à la fin de texte, obtenir début et fin de chaine
  i_tkn_debut = groupe_nominal[0]
  i_tkn_fin = groupe_nominal[-1]
  tkn_debut = l_tkn_sous_titres[i_tkn_debut]
  tkn_fin = l_tkn_sous_titres[i_tkn_fin]
  pos_debut = tkn_debut[2]
  pos_fin = tkn_fin[2] + len(tkn_fin[0])
  groupe_texte = txt_sous_titres[pos_debut:pos_fin]
  # il faut aussi substituer des _ aux blancs
  groupe_texte = groupe_texte.replace(' ', '_')

  return groupe_lemmes, groupe_tokens, groupe_texte
```

```
# 7.l
# récupération et comptage des groupes nominaux (et aussi des textes d'entités nommées)

d_groupes_lemmes = {}
d_groupes_tokens = {}
d_groupes_textes = {}
d_entites_nommees = {}

for ix_video, row in tqdm(df_videos.iterrows(), total=df_videos.shape[0]):

  txt_sous_titres = row['ss_titres']
  l_tkn_sous_titres = row['tkn_sous_titres']
  l_ent_sous_titres = row['ent_sous_titres']

  # groupes nominaux
  l_groupes_nominaux = groupes_nominaux_par_nlp(txt_sous_titres, nlp, matcher)
  for groupe_nominal in l_groupes_nominaux:
    groupe_lemmes, groupe_tokens, groupe_texte = reconstruit_groupe_nominal(groupe_nominal, l_tkn_sous_titres, txt_sous_titres)

    if groupe_lemmes not in d_groupes_lemmes: d_groupes_lemmes[groupe_lemmes] = 0
    d_groupes_lemmes[groupe_lemmes] += 1
    if groupe_tokens not in d_groupes_tokens: d_groupes_tokens[groupe_tokens] = 0
    d_groupes_tokens[groupe_tokens] += 1
    if groupe_texte not in d_groupes_textes: d_groupes_textes[groupe_texte] = 0
    d_groupes_textes[groupe_texte] += 1
  
  # entités nommées
  for entite_nommee in l_ent_sous_titres:
    entite_nommee_texte = entite_nommee[0]
    # ne garder que si l'entité nommée contient 2 ou plus tokens (tokénisation ultra simplifiée suffira)
    # remarque au vu des résultats, on pourrait éliminer les articles, 
    # car la_France, la_Chine, ne sont pas intéressants en tant que tels (non fait ici)
    l_tokens_entite_nommee_texte = entite_nommee_texte.split()
    if len(l_tokens_entite_nommee_texte) >= 2:
      entite_nommee_texte = entite_nommee_texte.replace(' ', '_')
      if entite_nommee_texte not in d_entites_nommees: d_entites_nommees[entite_nommee_texte] = 0
      d_entites_nommees[entite_nommee_texte] += 1
```

```
# 7.m
# conversion en listes triées selon la fréquence (en décroissant)
l_groupes_lemmes = [(groupe_lemmes, frequence) for groupe_lemmes, frequence in d_groupes_lemmes.items()]
l_groupes_lemmes.sort(key=lambda x: x[1], reverse=True)
l_groupes_tokens = [(groupe_tokens, frequence) for groupe_tokens, frequence in d_groupes_tokens.items()]
l_groupes_tokens.sort(key=lambda x: x[1], reverse=True)
l_groupes_textes = [(groupe_texte, frequence) for groupe_texte, frequence in d_groupes_textes.items()]
l_groupes_textes.sort(key=lambda x: x[1], reverse=True)
l_entites_nommees = [(entite_nommee_texte, frequence) for entite_nommee_texte, frequence in d_entites_nommees.items()]
l_entites_nommees.sort(key=lambda x: x[1], reverse=True)
```

###### Utilisation de Phrase de gensim

```
# 8.0
# préliminaire (si on n'est pas passé auparavant par 2.e à 2.i)

CATEGORIES_VIDES = ['ADP', 'AUX', 'CONJ', 'CCONJ', 'DET', 'PART', 'PRON',
                    'PUNCT', 'SCONJ', 'SYM', 'SPACE']
CATEGORIES_PLEINES = ['ADJ', 'ADV', 'INTJ', 'NOUN', 'NUM', 'PROPN', 'VERB', 'X']

def selectionne_lemme(lemme, pos, est_stopword):
  selection_lemme = ''
  if pos in CATEGORIES_PLEINES and est_stopword == False:
    selection_lemme = lemme
  return selection_lemme

def cree_l_items_sur_resultat_nlp(l_tokens_nlp, retour='lemme'):
  l_items = []
  for token in l_tokens_nlp:
    lemme = selectionne_lemme(token[3], token[4], token[7])
    if lemme != '':
      if retour == 'lemme': l_items.append(lemme)
      if retour == 'tout': l_items.append(token)
  return l_items

df_videos['l_textes_l_tokens'] = df_videos['tkn_sous_titres'].progress_apply(lambda x: cree_l_items_sur_resultat_nlp(x, retour='lemme'))

l_textes_l_tokens = df_videos['l_textes_l_tokens'].values.tolist()
```

```
# 8.a
# créée le modèle et calcule la liste des groupes pour une liste de textes (sous forme de listes de tokens, par exemple lemmatisés)

# scoring ({'default', 'npmi', function}, optional)
# min_count (float, optional) – Ignore all words and bigrams with total collected count lower than this value. 
# threshold (float, optional) – Represent a score threshold for forming the phrases (higher means fewer phrases). 
# A phrase of words a followed by b is accepted if the score of the phrase is greater than threshold. 
# Heavily depends on concrete scoring-function, 'default': > 0, 'npmi': between 0 and 1. 

def cree_modele_groupes_gensim(l_textes_l_tokens, methode='default', nb_min=5, seuil=None):
    
    # obtenir le modèle qui permet de remplacer les couples pertinents par leur concatenation
    # (comme on combine avec les éléments issus de l'analyse syntaxique, ne sera pas utilisé tel quel)
    if seuil is None:
        if methode == 'default': seuil_methode = 10
        if methode == 'npmi': seuil_methode = 0.5
    else:
        seuil_methode = seuil
    modele_groupes = Phrases(l_textes_l_tokens, scoring=methode, min_count=nb_min, threshold=seuil_methode,
                             delimiter=b'_')
    
    # obtenir les scores pour analyse des groupes de mots retenus
    x_groupes = modele_groupes.export_phrases(l_textes_l_tokens)
    l_groupes = [(groupe.decode("utf-8"), score) for groupe, score in x_groupes]
    # mettre aussi les groupes sous la forme lemme1_lemme2 au lieu de lemme1 lemme2
    l_groupes = [('_'.join(groupe.split()), score) for groupe, score in l_groupes] 
    # cette première liste est une liste des occurrences, la transformer en une liste de termes avec score et nb d'occurrences
    d_groupes = {}
    for (groupe, score) in l_groupes:
        if groupe not in d_groupes:
            d_groupes[groupe] = (score, 1)
        else:
            d_groupes[groupe] = (score, d_groupes[groupe][1] + 1)
    l_groupes = list(d_groupes.items())
    # tri par score, et non par fréquence
    l_groupes.sort(key=lambda x: x[1][0], reverse=True)

    # le modèle ne servira pas plus dans cette fonction, renvoyé pour utilisation dans d'autres contextes  
    return modele_groupes, l_groupes
```

```
# 8.b
methode_gensim='default'
nb_min_gensim=5
seuil_gensim=3
modele_groupes_stat, l_groupes_stat = cree_modele_groupes_gensim(l_textes_l_tokens, 
                                                    methode=methode_gensim, 
                                                    nb_min=nb_min_gensim,
                                                    seuil=seuil_gensim)
```

```
# 8.c
# pour npmi, min_count ne marche pas, l'appliquer ici
if methode_gensim == 'npmi':
  l_groupes_stat = [groupe for groupe in l_groupes_stat if groupe[1][1] >= nb_min_gensim]
```

```
# 8.i
# il peut être plus intéressant de montrer la relation entre score et fréquence, 
# pour voir quelle meilleure combinaison pour la sélection
def montre_distribution_score_frequence(l_groupes_stat):

  # en x, liste des fréquences (logarithmes de )
  l_freqs = [math.log(frequence, 10) for (groupe, (score, frequence)) in l_groupes_stat]
  # en y, liste des scores
  l_scores = [math.log(score, 10) for (groupe, (score, frequence)) in l_groupes_stat]
  fig = go.Figure()
  fig.add_trace(go.Scatter(x=l_freqs, y=l_scores, mode='markers'))
  titre = "Distribution fréquences / scores des groupes statistiques"
  xaxis_titre = "log fréquence"
  yaxis_titre = "log score"
  fig.update_layout(title=titre, xaxis_title=xaxis_titre, yaxis_title=yaxis_titre)
  fig.show()
```

```
# 8.k
# on peut aussi montrer un histogramme de la combinaison des deux
def montre_distribution_combinaison_score_frequence(l_groupes_stat, coeff_frequence=1.0):
  l_combis = [math.log(score, 10) + coeff_frequence * math.log(frequence, 10)\
              for (groupe, (score, frequence)) in l_groupes_stat]
  fig = go.Figure()
  fig.add_trace(go.Histogram(x=l_combis))
  titre = "Distribution combinaison scores.fréquences des groupes statistiques"
  xaxis_titre = "log score + %.1f * log fréquence" % coeff_frequence
  yaxis_titre = ""
  fig.update_layout(title=titre, xaxis_title=xaxis_titre, yaxis_title=yaxis_titre)
  fig.show()
```

##### Sélection des groupes statistiques les plus pertinents
###### Au vu du diagramme, faire combinaison score / fréquence, en privilégiant le score

```
# 8.m
# sélection sur combinaison du score et de la fréquence
def select_groupes_sur_combi(l_groupes_stat, val_min, coeff_frequence=1.0):
  l_sel_groupes = []
  for (groupe, (score, frequence)) in l_groupes_stat:
    if math.log(score, 10) + coeff_frequence * math.log(frequence, 10) > val_min:
      # on garde la fréquence, pour être en phase avec les données groupes nominaux
      l_sel_groupes.append((groupe, frequence))
  return l_sel_groupes
```
